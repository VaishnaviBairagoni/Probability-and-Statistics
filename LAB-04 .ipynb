```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/titanic_dataset_250-7.csv")
df.head()

#Histogram Analysis
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color='blue')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age among Passengers')
plt.show()

```

```python
#Scatterplot Exploration
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette={0: 'red', 1: 'green'})
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatterplot of Fare vs Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
#Bar Chart Visualization
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Class',hue='Class', palette='pastel')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Number of Passengers in Each Class')
plt.show()
```

```python
#Grouped Bar Chart
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Class', hue='Survived', palette={0: 'red', 1: 'green'})
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Count Across Passenger Classes')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

```

```python
# Analysis Answers
common_age_group = df['Age'].mode()[0]
print(f'The most common age group observed in the dataset is {common_age_group} years.')
```

```python
# Checking the trend between Age and Fare
correlation = df[['Age', 'Fare']].corr().iloc[0,1]
print(f'Correlation between Age and Fare: {correlation:.2f}')
if abs(correlation) > 0.2:
    print('There is a weak relationship between age and fare paid.')
else:
    print('There is no strong visible trend between age and fare paid.')
```

```python
# Identifying the class with the highest passengers
highest_class = df['Class'].value_counts().idxmax()
print(f'The passenger class with the highest number of passengers is Class {highest_class}.')

```

```python
# Survival variation across passenger classes
survival_rates = df.groupby('Class')['Survived'].mean()
print('Survival rates per class:\n', survival_rates)
```
