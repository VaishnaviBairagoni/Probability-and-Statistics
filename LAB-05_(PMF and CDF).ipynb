```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#1: Generate dataset
N = 1000
np.random.seed(42)
vehicle_counts = np.random.choice([0, 1, 2, 3, 4, 5], size=N,p=[0.1, 0.3, 0.25, 0.2, 0.1, 0.05])
```

```python
# Compute PMF
unique, counts = np.unique(vehicle_counts, return_counts=True)
pmf = counts / N
```

```python
# Compute CDF
cdf = np.cumsum(pmf)
```

```python
# Plot PMF
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(unique, pmf, width=0.5, color='b', alpha=0.7, edgecolor='black')
plt.xlabel("Number of Vehicles")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")
plt.xticks(unique);
```

```python
# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(unique, cdf, marker='o', linestyle='-', color='r')
plt.xlabel("Number of Vehicles")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.xticks(unique)

plt.tight_layout()
plt.show()
```

```python
#2: Binomial Distribution Analysis
N_gadgets = 100
p_defective = 0.05
```

```python
# i) Probability of exactly 10 defective gadgets
prob_10_defective = stats.binom.pmf(10, N_gadgets, p_defective)
print(f"Probability of exactly 10 defective gadgets: {prob_10_defective:.6f}")
```

```python
# ii) Probability of at least 15 defective gadgets
prob_at_least_15 = 1 - stats.binom.cdf(14, N_gadgets, p_defective)
print(f"Probability of at least 15 defective gadgets: {prob_at_least_15:.6f}")
```

```python
# iii) Expected number of defective gadgets
expected_defective = N_gadgets * p_defective
print(f"Expected number of defective gadgets: {expected_defective}")
```
