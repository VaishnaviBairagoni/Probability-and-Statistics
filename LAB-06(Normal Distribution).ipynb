```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Question 1: Generate normal distribution
mean = 50
std_dev = 5
size = 1000
data = np.random.normal(mean, std_dev, size)
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Plot Probability Density Function (PDF)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
pdf = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r', label='PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram and PDF of Normal Distribution')
plt.legend()
plt.show()

```

```python
# Question 2: Probability calculations
mean_test = 14
std_test = 2.5
sample_size = 1000

def normal_prob(lower, upper):
    return stats.norm.cdf(upper, mean_test, std_test) - stats.norm.cdf(lower, mean_test, std_test)

# I. Students scored between 14 and 2.5 (incorrect range, assuming meant 14 and 16.5)
p1 = normal_prob(14, 16.5) * sample_size

# II. Students scored below 12
p2 = stats.norm.cdf(12, mean_test, std_test) * sample_size

# III. Students scored above 19
p3 = (1 - stats.norm.cdf(19, mean_test, std_test)) * sample_size

print(f"Number of students scored between 14 and 16.5: {p1:.0f}")
print(f"Number of students scored below 12: {p2:.0f}")
print(f"Number of students scored above 19: {p3:.0f}")

```

```python
# Visualization of normal distribution
x_vals = np.linspace(mean_test - 4*std_test, mean_test + 4*std_test, 1000)
y_vals = stats.norm.pdf(x_vals, mean_test, std_test)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'r', label='PDF of Test Scores')
plt.fill_between(x_vals, y_vals, where=(x_vals >= 14) & (x_vals <= 16.5), color='g', alpha=0.3, label='Between 14 and 16.5')
plt.fill_between(x_vals, y_vals, where=(x_vals < 12), color='b', alpha=0.3, label='Below 12')
plt.fill_between(x_vals, y_vals, where=(x_vals > 19), color='orange', alpha=0.3, label='Above 19')
plt.xlabel('Test Scores')
plt.ylabel('Density')
plt.title('Normal Distribution of Test Scores')
plt.legend()
plt.show()
```
