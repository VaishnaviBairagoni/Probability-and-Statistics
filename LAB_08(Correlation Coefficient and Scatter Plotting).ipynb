```python
import numpy as np
import matplotlib.pyplot as plt

# Data
study_hours = np.array([2, 4, 6, 8, 1, 5, 7, 3, 9, 10])
exam_scores = np.array([50, 60, 70, 85, 40, 65, 78, 55, 90, 95])
sleep_hours = np.array([8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3])

# Scatter plots
plt.figure(figsize=(12, 4))

# Study Hours vs Exam Scores
plt.subplot(1, 3, 1)
plt.scatter(study_hours, exam_scores, color='blue')
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")

# Sleep Hours vs Exam Scores
plt.subplot(1, 3, 2)
plt.scatter(sleep_hours, exam_scores, color='red')
plt.title("Sleep Hours vs Exam Scores")
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Scores")

# Study Hours vs Sleep Hours
plt.subplot(1, 3, 3)
plt.scatter(study_hours, sleep_hours, color='green')
plt.title("Study Hours vs Sleep Hours")
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import pearsonr

# Pearson Correlation Coefficients
cor_study_exam, _ = pearsonr(study_hours, exam_scores)
cor_sleep_exam, _ = pearsonr(sleep_hours, exam_scores)
cor_study_sleep, _ = pearsonr(study_hours, sleep_hours)

# Print correlation values
print(f"Pearson Correlation (Study Hours vs Exam Scores): {cor_study_exam}")
print(f"Pearson Correlation (Sleep Hours vs Exam Scores): {cor_sleep_exam}")
print(f"Pearson Correlation (Study Hours vs Sleep Hours): {cor_study_sleep}")
```
